## A que circuito electoral pertencen los radios censales?

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import geopandas as gpd

from unidecode import unidecode
```

```python
# Cargar circuitos (BA)
circuitos_BA = gpd.read_file('./../EleccionesARG/data/circuitos/clean/clean.shp').reset_index(drop = True)
circuitos_BA.index.name = 'ix_circuitos'

# Cargar radios
radios = gpd.read_file('./Radios_censales/radios_w_geometry.shp').reset_index(drop = True)
radios.index.name = 'ix_radios'
# dptos info
DPTO = pd.read_csv('./../EleccionesARG/data/censo_geo/DPTO.csv', ';', dtype=str)[['DPTO', 'NOMDPTO']]
radios['DPTO'] = radios.LINK.str[1:5]
radios = radios.merge(DPTO, on = 'DPTO')
radios['dpto_merge'] = [unidecode(row) for row in radios['NOMDPTO'].str.lower()]
radios['dpto_merge'] = radios['dpto_merge'].str.replace('25', 'veinticinco').str.replace('9 ', 'nueve ')
# Seleccionar BA
radios_BA = radios.loc[radios.PROVINCIA.isin(['BUENOS AIRES', 'CIUDAD DE BUENOS AIRES'])]
```

```python
radios_BA.drop('geometry', axis = 1).reset_index().nunique()
```

```python
circuitos_BA.drop('geometry', axis = 1).reset_index().nunique()
```

```python
# Radios de un depto en particular
radios_BA_sample = radios_BA.loc[radios_BA.DPTO == '2012']
circuitos_BA_sample = circuitos_BA.loc[circuitos_BA.dpto_merge == 'comuna 12']
```

```python
from shapely.geometry import Polygon

polys1 = radios_BA_sample[['geometry']]
polys2 = circuitos_BA_sample[['geometry']]

df1 = gpd.GeoDataFrame(polys1)
df2 = gpd.GeoDataFrame(polys2)

import matplotlib.pyplot as plt
%matplotlib inline

fig, axs = plt.subplots(1, 3, figsize = (15, 6))

for i, gdf in enumerate([circuitos_BA_sample, radios_BA_sample, gpd.overlay(df1, df2, how='identity')]):
    ax = axs[i]
    gdf.reset_index(drop = True).reset_index().plot('index', ax = ax)
plt.show()
```

```python
radios_BA.dpto_merge.unique()[:10]
```

```python
resultado_dptos = []

for dpto in radios_BA.dpto_merge.unique():
    # Radios de un depto en particular
    radios_BA_dpto = radios_BA.loc[radios_BA.dpto_merge == dpto]
    circuitos_BA_dpto = circuitos_BA.loc[circuitos_BA.dpto_merge == dpto]
    
    polys1 = radios_BA_dpto[['geometry']]
    polys2 = circuitos_BA_dpto[['geometry']]

    # ver que radios se superponen con que circuitos y acomodar
    intersection = gpd.sjoin(gpd.GeoDataFrame(polys1), gpd.GeoDataFrame(polys2), how = 'left', op = 'intersects')
    intersection = intersection.reset_index().rename(columns = {'index': 'ix_radios', 'index_right': 'ix_circuitos'})
    intersection = intersection.merge(polys2.reset_index(), on = 'ix_circuitos')

    # Computar el poligono de interseccion...
    gdf1 = gpd.GeoDataFrame(intersection[['geometry_x']].rename(columns = {'geometry_x': 'geometry'}))
    gdf2 = gpd.GeoDataFrame(intersection[['geometry_y']].rename(columns = {'geometry_y': 'geometry'}))
    intersection['geometry'] = gdf1.intersection(gdf2)

    # para ver que porcion del area del radio esta en dicho circuito.
    intersection['area_radio'] = gpd.GeoSeries(intersection['geometry_x']).area
    intersection['area_int'] = gpd.GeoSeries(intersection['geometry']).area
    intersection['pertenencia'] = intersection.area_int/intersection.area_radio

    # Desestimar cuando la interseccion es menor al 1% y redondear.
    intersection = intersection.loc[intersection['pertenencia'] > 0.01].round(3)
    
    intersection['dpto_merge'] = dpto
    
    # guardar el resultado de este dpto y pasar al siguiente
    resultado_dptos += [intersection[['ix_radios', 'ix_circuitos', 'pertenencia', 'dpto_merge']]]
```

```python
resultado = pd.concat(resultado_dptos).reset_index(drop = True)
resultado.to_csv('./interseccion_radio_circu.csv')
```
