```python
import matplotlib.pyplot as plt
%matplotlib inline

import geopandas as gpd

pxlocdatos_1 = gpd.read_file("./Radios_censales/Buenos_Aires_con_datos.shp")
pxlocdatos_1.columns = [c.lower() for c in pxlocdatos_1.columns]
pxlocdatos_2 = gpd.read_file("./Radios_censales/cabaxrdatos.shp").rename({'TOT_POB': 'totalpobl'}, axis = 1)
pxlocdatos_2.columns = [c.lower() for c in pxlocdatos_2.columns]

pxlocdatos = pd.concat([pxlocdatos_1, pxlocdatos_2]).dropna(axis = 1)
# 
# pxlocdatos = gpd.read_file("./Radios_censales/Santa_Fe_con_datos.shp")
# pxlocdatos.columns = [c.lower() for c in pxlocdatos.columns]

pxlocdatos.sample(3)
```

```python
import pandas as pd

# datos = gpd.GeoDataFrame(pd.read_csv('./PERSONA-P03.csv'))

datos = gpd.GeoDataFrame(pd.read_csv('./PERSONA-P05.csv'))
datos['nacidos_exterior'] = datos.iloc[:, 2]/datos.iloc[:, 3]

datos.rename(columns={'radio': 'link', 'TOTAL': 'totalpobl'}, inplace=True)
datos['link'] = datos['link'].astype(str).str.zfill(9)

datos = pxlocdatos.merge(datos)

datos["area_km2"] = 10**-6 * datos['geometry'].area
datos['densidad'] = datos['totalpobl'] / datos["area_km2"]
datos['densidad_nacidos_exterior'] = datos['nacidos_exterior'] / datos["area_km2"]
```

```python
datos['densidad']
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
f, ax = plt.subplots(1, figsize=(25, 25))

# datos.plot(axes = ax, column = 'densidad', cmap='rainbow', edgecolors = 'None', vmin = 0, vmax = 25000, alpha = 0.5)
datos.plot(axes = ax, column = 'densidad_nacidos_exterior', cmap='rainbow', edgecolors = 'None', vmin = 0, vmax = 30, alpha = 0.5)

ax.set_xlim(4170000, 4210000)
ax.set_ylim(6130000, 6170000)
```

```python
plt.figure(figsize = (15, 5))
plt.hist(datos['nacidos_exterior'].dropna().values, 100, normed='True')
plt.show()
```
